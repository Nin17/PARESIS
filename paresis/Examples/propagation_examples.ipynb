```python
from propagator_2 import Fresnel
from samples_2 import Sphere, Ellipsoid, Cone, SuperQuadric
import numpy as np
import matplotlib.pyplot as plt
```

```python
test = Cone(radii=[5,15], height=6,pixel_size=1, dimensions = (15,17), position=np.array([0,0]))
test.show_sample()
a = Fresnel(test.thickness, 10)
a.show_propagation([2,4], 0.0001, 1e-6)
plt.show()
```

```python
test = Sphere(radius=2, pixel_size=2, dimensions = (5,7), position=np.array([0,0])).thickness
a = Fresnel(test, 10)
a.show_propagation([2], 0.0001, 1e-6)
plt.show()
```

```python
test = Sphere(radius=1, pixel_size=1, dimensions = (11,11), angle=0, position=np.array([0,0])).thickness
a = Fresnel(test, 10)
plt.figure()
plt.title('Input')
plt.imshow(test)
plt.colorbar()
PARAMS = [2, 0.0001, 1e-6]
PARAMS_ME = [[2], 0.0001, 1e-6]
plt.figure()
plt.title('Propagate')
plt.imshow(np.real(a.propagate(*PARAMS_ME))[0]-test)
plt.colorbar()
plt.figure()
plt.title('Propagate_Laurene')
plt.imshow(np.real(a.propagate_laurene(*PARAMS))-test)
plt.colorbar()
plt.figure()
plt.title('Propagate - Laurene')
plt.imshow(np.real(a.propagate_laurene(*PARAMS))-np.real(a.propagate(*PARAMS_ME))[0])
plt.colorbar()

```
